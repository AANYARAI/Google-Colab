```python
import pandas as pd
```

```python
d= pd.read_csv('Salary_Data.csv')
print(d)
```

```python
d.info()
print(d.isnull().sum())
```

```python
d = d.drop_duplicates()
print(d)
```

```python
d=pd.concat([d,d.sample(frac=1,replace=True)])
print(d)
```

```python
d=d.sample(frac=1)
d
```

```python
import matplotlib.pyplot as plt
```

```python
d['Age'] = d['Age'].fillna(d['Age'].mean())
d['Years of Experience'] = d['Years of Experience'].fillna(d['Years of Experience'].median())
d['Gender'] = d['Gender'].fillna(d['Gender'].mode()[0])
d
```

```python
d.dropna()
d
```

```python
d['Experience_Level'] = pd.cut(
    d['Years of Experience'],
    bins=[0,3,7,15,50],
    labels=['Junior','Mid','Senior','Expert']
)
d
```
